```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df.columns
```

```python
plt.title('age vs sal')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased']);
```

# Normalization

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# create an object
scaler = MinMaxScaler()
```

```python
# apply scaling
x = df[['Age', 'EstimatedSalary']]

x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled[:,0].min(), x_scaled[:,0].max()  # age
```

```python
x_scaled[:,1].min(), x_scaled[:,1].max()  # salary
```

```python
plt.title('age vs sal')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x_scaled[:,0], x_scaled[:,1], c = df['Purchased']);
```
