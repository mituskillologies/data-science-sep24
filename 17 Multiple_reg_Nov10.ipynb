```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mtcars.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x, y)
```

```python
lr.score(x, y)
```

```python
disp = 210
hp = 100
wt = 3
```

```python
lr.predict([[disp, hp, wt]])
```
