```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([12, 18, 24, 28, 34, 40, 45, 51, 59, 64])

y = np.array([20, 25, 29, 33, 38, 43, 49, 55, 59, 61])

z = np.array([67, 61, 57, 49, 47, 41, 36, 30, 27, 20])
```

# Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--');
plt.grid()

# marker => >, <. ^, H, *, o
```

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second Value')
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--');
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--');
plt.plot(x, z, marker = 'o', mfc = 'orange', ls = '--', color = 'black');
plt.grid()

# marker => >, <. 
```

#### subplot

```python
plt.figure(figsize=(16,9))

plt.subplot(2, 2, 1)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--');
plt.grid()

plt.subplot(2, 2, 2)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second Value')
plt.plot(x, z, marker = 'o', mfc = 'red', ls = '--');
plt.grid()
```

# Bar Plot

```python
courses = ['java', 'python', 'kotlin', 'dart', 'c++']
students = [50, 61, 58, 41, 30]
```

```python
plt.title('Student Enrollment')
plt.xlabel('Course Name')
plt.ylabel('Students enrolled')
plt.bar(courses, students, color = 'black', width = 0.3);
plt.grid()

# load student.csv
# plot ar graph for student name and marks scored
```

```python
df = pd.read_csv('student.csv')
df.head(2)
```

```python
plt.figure(figsize=(16,9))
plt.title('student marks')
plt.xlabel('name')
plt.ylabel('marks')
plt.bar(df['name'], df['marks'], color = 'black')
plt.grid();
```

# Scatter plot

```python
plt.scatter(x, y, marker = '*')
plt.grid()
```

# Pie Chart

```python
plt.title('Pie chart')
plt.pie(students, labels = courses, autopct = '%.2f%%',
       shadow = True);

# plot pie chart, combined line plot, bar plot, scatter plot in a single
# figure
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--');
plt.plot(x, z, marker = 'o', mfc = 'orange', ls = '--', color = 'black');
plt.grid()

plt.subplot(2,2,2)
plt.title('student marks')
plt.xlabel('name')
plt.ylabel('marks')
plt.bar(df['name'], df['marks'], color = 'black')
plt.grid();

plt.subplot(2,2,3)
plt.scatter(df1['Age'], df1['EstimatedSalary'],
           c = df1['Purchased']);
plt.grid()

plt.subplot(2,2,4)
plt.title('Pie chart')
plt.pie(students, labels = courses, autopct = '%.2f%%',
       shadow = True);

plt.savefig('MyPlot.png')
```
